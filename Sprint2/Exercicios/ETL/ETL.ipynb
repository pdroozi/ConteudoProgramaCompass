```python
# 3.1 Etapa 1 - Leitura de arquivo CSV e Ator/Atriz com mais filmes

dados = []

with open('actors.csv', 'r', encoding='utf-8') as arq:
    topicos = next(arq).strip().split(',')
    topicos = [c.strip().strip('"') for c in topicos]

    for linha in arq:
        if not linha:
            continue
        else:
            linha = linha.strip()
    

        valores_linha = []
        valores_campo = ''
        campo_com_aspas = False

        for x in linha:
            if x == '"':
                campo_com_aspas = not campo_com_aspas
            elif x == ',' and not campo_com_aspas:
                valores_linha.append(valores_campo.strip().strip('"'))
                valores_campo = ''
            else:
                valores_campo += x

        valores_linha.append(valores_campo.strip().strip('"'))
        dados.append(valores_linha)

filmes = 0
ator_atriz = ''

for x in dados:
    ator = x[0]  
    num_filmes = int(x[2]) 
    if num_filmes > filmes:
        filmes = num_filmes
        ator_atriz = ator

print('O artista que mais atuou em filmes é:', ator_atriz + ' com', filmes, 'filmes.')
```

```python
# 3.2 Etapa 2 - Média de receita dos filmes

dados = []

with open('actors.csv', 'r', encoding='utf-8') as arq:
    topicos = next(arq).strip().split(',')
    topicos = [c.strip().strip('"') for c in topicos]

    for linha in arq:
        if not linha:
            continue
        else:
            linha = linha.strip()
    

        valores_linha = []
        valores_campo = ''
        campo_com_aspas = False

        for x in linha:
            if x == '"':
                campo_com_aspas = not campo_com_aspas
            elif x == ',' and not campo_com_aspas:
                valores_linha.append(valores_campo.strip().strip('"'))
                valores_campo = ''
            else:
                valores_campo += x

        valores_linha.append(valores_campo.strip().strip('"'))
        dados.append(valores_linha)



soma_bilheteria = 0
total_filmes = 0

for linha in dados:
    gross = float(linha[-1])  # Última coluna
    soma_bilheteria += gross
    total_filmes += 1

if total_filmes > 0:
    media = soma_bilheteria / total_filmes
    media_arredondada = round(media, 2)
    print(f'A média de receita dos filmes é: {media_arredondada} milhões de dólares.')
```

```python
# 3.4 Etapa 4 - Aparições de filmes no CSV

dados = []

with open('actors.csv', 'r', encoding='utf-8') as arq:
    topicos = next(arq).strip().split(',')
    topicos = [c.strip().strip('"') for c in topicos]

    for linha in arq:
        if not linha:
            continue
        else:
            linha = linha.strip()
    

        valores_linha = []
        valores_campo = ''
        campo_com_aspas = False

        for x in linha:
            if x == '"':
                campo_com_aspas = not campo_com_aspas
            elif x == ',' and not campo_com_aspas:
                valores_linha.append(valores_campo.strip().strip('"'))
                valores_campo = ''
            else:
                valores_campo += x

        valores_linha.append(valores_campo.strip().strip('"'))
        dados.append(valores_linha)


total= {}

for x in dados:
    filme = x[4]  
    if filme in total:
        total[filme] = total[filme] + 1
    else:
        total[filme] = 1

filmes_ordenados = sorted(total.items(),key=lambda x: (-x[1], x[0]))

for filme, quantidade in filmes_ordenados:
    if quantidade == 1:
        print(f'O filme {filme} aparece uma vez.')
    elif quantidade > 1:
        print(f'O filme {filme} aparece {quantidade} vezes.')
    else:
        print(f'O filme {filme} não aparece.')
    
```
