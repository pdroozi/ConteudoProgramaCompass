```python
import pandas as pd ### Ferramenta para manipulação de dados
import matplotlib.pyplot as plt ### Biblioteca para visualização de dados e gráficos

df_google_semDuplicados = pd.read_csv('../Desafio/Ex4.2.1/googleplaystore_semDuplicados.csv')

df_google_semDuplicados['Installs'] = (df_google_semDuplicados['Installs'].astype(str).str.replace('[+,]', '', regex=True))
df_google_semDuplicados['Installs'] = pd.to_numeric(df_google_semDuplicados['Installs'], errors='coerce')

top5Apps = df_google_semDuplicados.sort_values(by='Installs', ascending=False).head(5)
top5Apps
top5Apps.to_csv('../Desafio/Ex4.2.2/top5Apps.csv', index=False)
```

```python
import pandas as pd ### Ferramenta para manipulação de dados
import matplotlib.pyplot as plt ### Biblioteca para visualização de dados e gráficos

df_google_semDuplicados = pd.read_csv('../Desafio/Ex4.2.1/googleplaystore_semDuplicados.csv')

df_google_semDuplicados['Installs'] = (df_google_semDuplicados['Installs'].astype(str).str.replace('[+,]', '', regex=True))
df_google_semDuplicados['Installs'] = pd.to_numeric(df_google_semDuplicados['Installs'], errors='coerce')

top5Apps = df_google_semDuplicados.sort_values(by='Installs', ascending=False).head(5)

plt.figure(figsize=(8, 4))
plt.barh(top5Apps['App'], top5Apps['Installs'], color='green', edgecolor='black')
plt.xlabel('Número de Instalações (em bilhões)')
plt.ylabel('Aplicativos')
plt.title('Top 5 Apps com Mais Instalações')
plt.gca().invert_yaxis()  # Inverte o eixo y para mostrar o maior valor no topo
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig('../Desafio/Ex4.2.2/top5_apps_instalacoes.png', dpi=300, bbox_inches='tight')
plt.show()
```

O objetivo da atividade foi realizar a construção de um gráfico de barras, que evidencie os 5 apps mais instalados no dataframe, para isso, foi realizado um filtro de ordenação pelos numeros de instalações, no formato decrescente, limitado a 5 linhas, buscando então os top5 apps. ### 4.2.3

```python
import pandas as pd ### Ferramenta para manipulação de dados
import matplotlib.pyplot as plt ### Biblioteca para visualização de dados e gráficos

df_google_semDuplicados = pd.read_csv('../Desafio/Ex4.2.1/googleplaystore_semDuplicados.csv')

qtd = df_google_semDuplicados['Category'].value_counts()

top7= qtd.head(7)
outros = qtd[7:].sum()
total = pd.concat([top7, pd.Series({'Outros': outros})])

explode[7] = 0.05

plt.figure(figsize=(8, 6))
plt.pie(total, labels=total.index, explode=explode, autopct='%1.1f%%', startangle=140, colors=plt.cm.tab10.colors[:9], shadow=False)
plt.title('Distribuição de Apps por Categoria (Top 7 + Outros)')
plt.axis('equal')
plt.legend([*top7.index, 'Outros'], loc='upper right', bbox_to_anchor=(1.1, 1))
plt.tight_layout()
plt.savefig('../Desafio/Ex4.2.3/distribuicao_categorias_app.png', dpi=300, bbox_inches='tight')
plt.show()
```

O objetivo da atividade foi realizar a construção de um gráfico de pizza/torta, que evidencie as categorias de apps mais recorrentes, para isso, foi realizado um filtro de contagem pelos numeros de ocorrencias, no formato decrescente, limitado a 7 linhas (por escolha própria), buscando então as top7 categorias mais frequentes. O resto das categorias foram agrupadas em uma categoria Outros, para facilitar a interpretação. Após isso, o gráfico foi contruído com auxílio do matplotlib.

```python
import pandas as pd ### Ferramenta para manipulação de dados
import matplotlib.pyplot as plt ### Biblioteca para visualização de dados e gráficos

df_google_semDuplicados = pd.read_csv('../Desafio/Ex4.2.1/googleplaystore_semDuplicados.csv')

def conversaoParaMB(tamanho):
    if tamanho.endswith('M' or tamanho.endswith('m')):
        return float(tamanho[:-1])
    elif tamanho.endswith('K' or tamanho.endswith('k')):
        return float(tamanho[:-1]) / 1000
    else:
        return None

df_google_semDuplicados['Size'] = df_google_semDuplicados['Size'].apply(conversaoParaMB)

df_google_semDuplicados['Last Updated'] = pd.to_datetime(df_google_semDuplicados['Last Updated'], errors='coerce')

df_google_semDuplicados['Year'] = df_google_semDuplicados['Last Updated'].dt.year

tamanhoMedio= df_google_semDuplicados.groupby('Year')['Size'].mean()

plt.figure(figsize=(10, 5))
tamanhoMedio.plot(marker='o', color = 'blue')
plt.xticks(rotation=45)
plt.title('Tamanho Médio dos Apps por Ano')
plt.xlabel('Ano')
plt.ylabel('Tamanho Médio (MB)')
plt.grid(True)
plt.savefig('../Desafio/Ex4.2.8/tamanho_medio_apps.png', dpi=300, bbox_inches='tight')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df_google_semDuplicados = pd.read_csv('../Desafio/Ex4.2.1/googleplaystore_semDuplicados.csv')

df_google_semDuplicados['Last Updated'] = pd.to_datetime(df_google_semDuplicados['Last Updated'], errors='coerce')

df_google_semDuplicados['Year'] = df_google_semDuplicados['Last Updated'].dt.year

conteudoPorAno = pd.crosstab(df_google_semDuplicados['Year'],df_google_semDuplicados['Content Rating'])

conteudoPorAnoProporcao = conteudoPorAno.divide(conteudoPorAno.sum(axis=1), axis=0)

plt.figure(figsize=(12, 10))
conteudoPorAnoProporcao.plot(kind='barh', stacked=True, colormap='tab10')
plt.title('Proporção de Avaliação do Conteúdo por Ano')
plt.xlabel('Proporção de Apps')
plt.ylabel('Ano')
plt.legend(title='Avaliação do Conteúdo')
plt.savefig('../Desafio/Ex4.2.8/conteudoPorAno.png', dpi=300, bbox_inches='tight')
plt.tight_layout()
plt.show()

```

O objetivo dessa foi utilizar duas novas exibições gráficas não utilizadas anteriormente, as escolhidas foram 'LinePlot' e 'Barh Plot'

A primeira representão gráfica foi usada pra gerar um gráfico de linha que representa a evolução do tamanho médio dos aplicativos ao longo dos anos, já que na coluna size não há um valor definido, temos que criar um padrão, e o escolhido foi Megabyte, portanto, uma função foi criada com a tarefa de percorrer todas as linhas da tabela size, converter valores em float para comparação, onde caso ele termina com 'M', ele elimina a letra, e retorna o valor, caso ele termine com 'K', ele elimina a letra, e retorna o valor divididor por 1000, já que está sendo convertido Kilobytes para Megabyes. Com o filtro feito, realizamos a extração do Ano dos apps, através da função 'to_datetime' e depois 'dt.Year'. O resultado foi agrupado através do 'groupby' e depois utilizado para criaçaão do gráfico. 

A segunda representação gráfica foi usada pra gerar um gráfico de níveis de barra horizontais, que representa a evolução da proporção de categorias dos apps ao longo dos anos, realizamos como no exemplo acima a extração do ano, utilizando as mesmas funções, e o agrupamento das mesmas, após isso pelo 'sum' foi somado todos os apps de cada ano, e pelo 'divide' nós o dividimos pelo total, gerando a proporção do app de cata categoria em relação a cada ano, essas proporção foram usadas para gerar o gráfico como mostrado acima.
